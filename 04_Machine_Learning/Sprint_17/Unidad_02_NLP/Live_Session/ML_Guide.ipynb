<p>
    
Los strings se almacenan internamente en un conjunto de bytes, caracter a caracter. Esta operacin es lo que se conoce como *encoding*, mientras que pasar de bytes a string sera *decoding*. Bien, y eso en qu nos afecta? Dependiendo del encoding, se suelen almacenar los caracteres en un espacio de bits de 0 a 255, es decir, en esa combinacin de bits tienen que entrar todos los caracteres del lenguaje.

El problema es que en toda esa combinacin de bits no entran todos los caracteres del planeta, por lo que dependiendo del encoding que usemos, una combinacin de bits significar una cosa u otra. Por ejemplo, una A mayuscula ser lo mismo en el encodig europeo que en el americano, pero los bits reservados para representar una , en el encodig americano se traduce en otro caracter.

#### Me vale esta muestra para entrenar al modelo?
Es muy sencillo cuando tenemos un dataset cerrado, que viene de un concurso de Kaggle, pero en un caso real eso se cumple pocas veces. Hay muchas bases de datos en la empresa, por no hablar de todos los sitios externos (web scraping o APIs) de donde podemos sacar datos. Cmo sabemos que tenemos los datos suficientes para montar un modelo? Si vamos a obtener nuevos, a qu fuentes acudo? cmo es la calidad de estos datos?

1. **Volumen**: Lo primero, necesitamos un buen volumen de datos. Menos de mil observaciones suele ser escaso para entrenar y testar un modelo de machine learning.
2. **Calidad**: la calidad de los datos siempre que mejor que cantidad. Es mejor encontrar unas pocas features predictivas, cuyos datos sean fiables, que una gran cantidad de features que no aporten nada al modelo. Asegrate de que los datos conseguidos son buenos y no estn manipulados, ni modificados por otros integrantes de la empresa. Y de ser as, si vas a utilizarlos piensa que cuando realices predicciones, las entradas de tu modelo tendrn que ser esos mismos datos modificados. 

3. **Caso de uso**: piensa en el problema de negocio y planteate qu variables podran ser predictivas, y si es factible conseguir esos datos.
4. **Poblacin**: asegrante de que la poblacin/muestra utilizada para entrenar se asemeje a la poblacin con la que hars predicciones. Por ejemplo, si creas un modelo de tratamiento de imgenes, con el que predigas si unos pulmones tienen cncer o no, tu modelo no tendr un buen performance si lo entrenas con muestras de pulmones asiticos y pretendes predecir muestras de pulmones caucsicos. O si entrenas solo con pulmones de mujeres e intentas predecir sobre pulmones de hombres.
5. **Fuentes externas**: si tienes tiempo planteate acudir a fuentes externas a la empresa, mediante APIs, datasets de kaggle, pginas del gobierno... Por otro lado, no incluyas datos en el entrenamiento que luego no vayas a conseguir para las predicciones. Por ejemplo, si consigues una muestra concreta de datos que pubic una empresa, y ya no se van a publicar ms, cuando vayas a hacer predicciones, no vas a poder contar con esos datos. Por ltimo, piensa tambin que cuantas ms fuentes externas, ms dependencias tendr tu modelo para realizar las predicciones y quiz no sea sencillo conseguir ese dato.

6. **Crea tus propios datos**: no tienes datos? "*invntatelos*". Si necesitas crear un software de reconocimiento de imgenes para saber si alguien lleva gafas o no, saca fotos de amigos o familiares y utilizalas para el modelo. Otra opcin es realizar encuestas.

Es imposible saber a priori qu modelo y qu configuracin es la que mejor se ajusta a nuestros datos, por lo que tendremos que probar varias combinaciones (entrenar con train) y despus probar con test qu modelo da menos errores, y as elegir.

<details>

<p>
Los dos primeros argumentos son features y target. Dos cosas importantes. Lo primero, tienen que coincidir en dimensin, y lo segundo acurdate de quitar el target en el conjunto de las features. Despus especificamos la proporcin de los datos que reservaremos para test (0-1).

Realmente hay que dividir los datos en train, validation y test (ver imagen abajo). El conjunto de validation lo utilizamos para escoger modelo, y el de test para comprobar que nuestro modelo escogido generaliza bien. De momento vamos a preocuparnos de **guardar unos datos para test, y no los tocaremos hasta el final. No podemos contaminar train con los datos de test**. Esto supone seguir ciertas normas:

1. **Scalers**: por ejemplo, si tenemos un StandardScaler en train, usar ese mismo StandardScaler en test (ver ejemplo abajo).
2. **Outliers**: Me temo que los outliers no los voy a poder eliminar en test ya que en un problema real, podria entrar un outlier como input del modelo una vez hayamos hecho la puesta en produccin. Cuidado con eliminar registros si estamos en una competicin de Kaggle porque la cantidad de muestras que van a test no puede variar.
3. **Missings**: si aplicbamos la media/mediana/moda en train, aplicar esa misma mtrica en test. No apliques la media de test en test.
4. **Feature reduction**: eliminar las features que quitbamos en train.
5. **Feature engineering**: mismos clculos que en train.

**NOTA**: si se trata de un concurso de Kaggle, en el que ya te dan el conjunto de test aparte, no hace falta que dividas tu train para obtener un nuevo conjunto de test, sobre todo si tienes pocos datos. Utiliza todo el conjunto de train en el *cross validation* para escoger el mejor modelo, y despus el test proporcionado por Kaggle para realizar una submission.

# 4. Target
Tenemos que echar un pequeo vistazo al target ya que hay ciertas casusticas que no queremos tener, como por ejemplo un dataset desbalanceado o un target asimtrico. 

|Problema ML| Caso deseado | Caso no deseado |
|---|----|----|
|Clasificacin| Dataset balanceado| Dataset desbalanceado|
|Regresin| Distribucin normal| Distribucin asimtrica|

#### Clasificacin
Si intentamos predecir si una persona va a padecer un cncer o no, y entrenamos un modelo en el que el 97% de los casos no ha padecido cncer, si el modelo predice siempre que no tiene cncer, conseguir un 97% de precisin. Y no es lo que estamos buscando. Queremos profundizar ms en ese 3%, y que el modelo sepa discriminar bien entre una clase u otra. **Esto es lo que se conoce como dataset desbalanceado**. Lo ideal es que el dataset est lo ms equilibrado posible. Cmo lidiar con ello?
1. **Cambiando la mtrica**: recall, f1-score o ROC curve (ver apartado de mtricas).
2. **Consiguiendo ms datos**: Parece una tontera pero muchas veces ni nos lo planteamos
3. **Resampling**: poner ms copias del target con menor cantidad (over-sampling) o al revs (under-sampling), quitar datos del caso con mayor cantidad de datos. Under-sampling est bien cuando tenemos muchos datos, mientras que over-sampling lo tendremos que considerar cuando sean escasos. Prueba varios ratios.
4. **Generar datos sintticos**: existen funciones de sklearn que son capaces de generar nuevos datos, como SMOTE (Synthetic Minority Over-sampling Technique)
5. **Prueba nuevos algoritmos**: los rboles de decisin suelen funcionar bien en problemas desbalanceados.
6. **Penalizacin en los modelos**: modelos que presten ms atencin a la clase minoritaria.
7. **Weights**: Es posible aplicarle un peso diferente a cada clase, para que est ms balanceadas, para ello usamos el parmetro *class_weight*.

````python
def data_report(df):
    # Sacamos los NOMBRES
    cols = pd.DataFrame(df.columns.values, columns=["COL_N"])

    # Sacamos los TIPOS
    types = pd.DataFrame(df.dtypes.values, columns=["DATA_TYPE"])

    # Sacamos los MISSINGS
    percent_missing = round(df.isnull().sum() * 100 / len(df), 2)
    percent_missing_df = pd.DataFrame(percent_missing.values, columns=["MISSINGS (%)"])

    # Sacamos los VALORES UNICOS
    unicos = pd.DataFrame(df.nunique().values, columns=["UNIQUE_VALUES"])
    
    percent_cardin = round(unicos['UNIQUE_VALUES']*100/len(df), 2)
    percent_cardin_df = pd.DataFrame(percent_cardin.values, columns=["CARDIN (%)"])

    concatenado = pd.concat([cols, types, percent_missing_df, unicos, percent_cardin_df], axis=1, sort=False)
    concatenado.set_index('COL_N', drop=True, inplace=True)


    return concatenado.T
        
data_report(titanic_df)
````

## 11. Missings
Suponen un problema ya que la mayora de los modelos no saben tratar los missings y por tanto tendremos que inferir sus valores. Por aclarar, un missing no es un 0, ni un string vaco, ni un False. Es un hueco en los datos, el vaco, nos falta un dato en una de las features. Eso es un missing.

#### Por qu aparecen missings?
* **Missings en la extraccin de los datos**: estos se suelen identificar rapido, por ejemplo, que no se ha leido bien un CSV, y todos los valores se los tome como uno solo.
* **Missings completamente aleatorios**: la probabilidad de que haya un valor missing es la misma para todas las observaciones.
* **Missings aleatorios**: aleatoriamente hay missings en una columna. Por ejemplo, datos de edades de una pagina web, vemos que hay ms missings de mujeres que de hombres.
* **Missings que dependen de los inputs**: Por ejmplo, en un estudio mdico, si un diagnstico es dudoso, hay una alta probabilidad de ser descartado. Cuando estos missings se les pone a "Dudoso", dejan de ser valores aleatorios.
* **Missings que dependen de su propio valor**: por ejemplo, ingresos de las personas, los que tienen mucho o muy poco, se tender a no poner el valor.
* **Missings en tratamiento de datos**: puededn aparecer missings debido a una incorrecta lectura o escritura, tratamiento de los datos y merge con otras tablas.

#### Mtodos para tratar los missings
* **Imputacin sobre caso similar**: tengo en cuenta otra variable, por ejemplo si tengo la altura de alumnos de una clase, podria calcular los missings teniendo en cuenta si son chicos o chicas. O si hay missings de densidad de poblacin, pero tengo la poblacin y el rea, podra recalcularlo. Esta opcin es lo ideal, razonando los missings con los propios datos.
* **Borrado**: Aqu hay dos opciones: si hay un missing en una columna, nos cargamos toda la fila. El borrado se realiza cuando tenemos pocos missings y muchas observaciones. De lo contrario, eliminaramos muchos datos.
* **Imputacin por valor concreto**: No es lo ms habitual, pero puede ocurrir que sepamos de antemano por qu hay missings en esa feature. Quiz todos los das hay problemas en la carga de la base de datos con el carcter , y cuando debera aparecer Espaa, en realidad est en missing. Este mtodo tambin se utiliza cuando hay missings en la variable categrica, le ponemos un nombre genrico y equivaldra a aadir una categora nueva a la feature.
* **Imputacin de Media/Mediana/Moda**: es el mtodo ms frecuente, sobretodo si tenemos poco tiempo. Para variables numricas imputamos normalmente la media, aunque si la variable tiene asimetra, suele dar mejores resultados la mediana. En el caso de que sea una variable categrica lo ms habitual es utilizar la moda.
* **Imputacion + flag**: es igual que el caso anterior, pero adems aadimos una nueva columna binaria por cada feature con missings imputados para indicarle al modelo que en ese lugar haba un missing.
* **Modelo**: es lo ms sofisticado. Dividimos el dataset en dos: uno con missings y el otro sin, que se usa para entrenar al modelo. Este mtodo se suele comportar bien, a no ser que no tengan mucha relacin los datos. Lo habitual es usar KNN, regresin logstica o lineal. El KNN rellena los missings teniendo en cuenta el resto de variables del dataset, buscando similitudes con registros que tengan las variables parecidas. Se puede usar para cualitativas y cuantitativas, las variables con muchos missings se pueden tratar fcilmente. Lo malo es que es muy lento y la eleccin del k-value es crtica.

## 13. Outliers
#### Qu es un outlier?
Se trata de un valor atpico dentro de nuestros datos. Un valor que se desva mucho de las mtricas estadsticas de centralidad (media, moda, mediana). En el anlisis exploratorio de datos suele ser algo a estudiar, el por qu tengo algunos valores atpicos en los datos. No obstante, en machine learning los outliers en los datos implican penalizaciones en los modelos, sobretodo los que trabajan con distancias y con Gradient Descent. Por tanto, hay que lidiar con ellos.

#### Cmo detecto los outliers?
Lo mejor es grficamente, aunque tambin puedes parametrizar una serie de umbrales en tus features, a partir de los cuales consideras que son outliers. Hay dos maneras de verlos grficamente, mediante anlisis:
1. **Univariante**
2. **Bivariante**
3. **Medidas estadsticas**: cunto me desvo de la media, o cunto de lejos estoy del IQR.

#### Qu grficas utilizo para visualizar los outliers?
Lo mejor son boxplots, histogramas, diagramas de densidad, scatter plots y count plots para categricas.

#### Qu tcnicas hay para detectar outliers?
1. **Grficamente**: Datos que se desvien mucho.
2. **Cuartiles**: se ve en un diagrama de caja. Datos que caigan fuera del rango +/- 1.5*IQR. Este 1.5 es muy restrictivo por lo que se suelen probar valores del 3 al 5. Depender mucho de cada feature.
3. **Desviacin estndar**: todo lo que caiga fuera de (media +/- N*std) de la variable. Normalmente N es un valor de 3 a 5.

Ten en cuenta siempre el contexto de negocio para considerar los outliers, no el puramente numrico. El propio negocio del dataset que ests analizando es el que define qu es un outlier y qu no.

Detectar grficamente es la tcnica ms habitual, pero si quieres hacerlo de una manera ms automatizada puedes probar con las otras dos opciones. Adems ten en cuenta que **los outliers siempre dependen del negocio** por lo que debers analizarlos bien antes de eliminarlos.

### 14.4 Escalados
Dependiendo del modelo que vayamos a utilizar, resultar til escalar las features. **Qu significa escalar una variable?**
Consiste en cambiar el orden de magnitud de la variable, conservando su distribucin. Estas transformaciones se aplican por separado a cada feature. Se suele calcular unas mtricas como media, desviacin estndar, y mediante una frmula aplicamos la transformacin a cada valor de la feature.

#### Por qu realizamos transformaciones?
Hay ciertas caractersticas que le vienen bien a los modelos. Tener todos los datos en la misma escala es una de ellas. Los modelos trabajan mejor y entrenan ms rpido cuando la escala de todos los datos es la misma. Esto lo vamos a conseguir con un StandardScaler o un MinMaxScaler. Otra de las caractersticas que deberan tener los modelos es tener datos con distribuciones gausianas. Esto no siempre es as ya que muchas veces las distribuciones son bimodales o asimtricas. Para solucionar esto aplicamos logaritmos, elevamos al cuadrado o al cubo (veremos en el siguiente apartado).

#### StandardScaler

<details>
<summary><b>Explicacin del StandardScaler</b></summary>
<p>
La estandarizacin (o Z-Score) consiste en reescalar una feature para conseguir las propiedades de una distribucion normal estndar: $\mu = 0$ and $\sigma = 1$.

Donde $\mu$ es la media y $\sigma$ la desviacin estndar. De esta forma cada valor de la feature se transforma mediante la frmula:

\begin{equation} z = \frac{x - \mu}{\sigma}\end{equation} 

No solo conseguimos que todas las variables se encuentren en la misma escala, sino que beneficia al proceso de entrenamiento, particularmente todos los algoritmos que dependan de distancias (como KNN) o los que utilicen Gradient Descent, consiguiendo una convergencia del algoritmo ms rpida.

**En qu algoritmos usaremos StandardScaler?** Este escalado va a venir bien en todos menos en los rboles, a los que no les afecta el escalado. Algoritmos de clustering tambin van a demandar un StandardScaler ya que buscan similitudes entre features, midiendo la distancia entre las mismas. PCA tambin.

**Debo estandarizar el target?** Lo primero, tendra que ser un problema de regresin. Aunque no se suele estandarizar. Se deja como est y las predicciones van en la escala del target. No obstante, se podra realizar dicha estandarizacin y despus aplicarle la inversa (ver cdigo abajo).

**Cmo aplico el escalado a train y test?**. Primero creo los objetos scaler con los datos de train (SOLO los de train), y aplica ese scaler tanto a train como a test. As no contamino el conjunto de train con los datos de test.
    
**Cambia la forma de la distribucin con este scaler?** NO. Si tenas una distribucin asimtrica, seguir siendo asimtrica pero con otra escala.
    
</p>
</details>

Por desgracia los modelos no son perfectos y siempre cometen cierto error. De no ser as, probablemente no sera necesario utilizar Machine Learning para solucionar el problema y tengamos que acudir a soluciones ms sencillas.

Por tanto, tendremos que definir una mtrica para evaluar el comportamiento del modelo por dos motivos:

2. Segundo, para presentar **resultados del modelo**.

### 16.1 Mtricas para clasificacin

##### Accuracy
La mtrica ms habitual. Calcula el % de acierto teniendo en cuenta todas las clases del algoritmo de clasificacin. Esta mtrica es muy fcil de entender, pero no profundiza en el % de acierto de cada clase, que puede ser algo interesante, dependiendo del problema que queramos tratar.

Accuracy = (TP + TN) / Total

##### Matriz de confusin
Confusin o *error matrix* es una tabla que describe el rendimiento de un modelo supervisado de Machine Learning en los datos de test, donde se desconocen los verdaderos valores. Se llama matriz de confusin porque hace que sea fcil detectar dnde el sistema est confundiendo dos clases.

* **True Positives (TP)**: cuando la clase real del punto de datos era 1 (Verdadero) y la predicha es tambin 1 (Verdadero)
* **Verdaderos Negativos (TN)**: cuando la clase real del punto de datos fue 0 (Falso) y el pronosticado tambin es 0 (Falso).
* **False Positives (FP)**: cuando la clase real del punto de datos era 0 (False) y el pronosticado es 1 (True).
* **False Negatives (FN)**: Cuando la clase real del punto de datos era 1 (Verdadero) y el valor predicho es 0 (Falso).

##### Recall o Sensibilidad
Los positivos que he clasificado bien vs todos los positivos que haba. Mtrica que se utiliza cuando queremos ahcer foco en minimizar los FN (Falsos Negativos). Claro ejemplo puede ser un test de COVID. No me importa tanto que haya FP, ya que las consecuencias son aislamientos preventivos, mientras que tener FN, es decir, personas con COVID, cuyo resultado del test es negativo, si es grave ya que pueden producir ms contagios.

Recall = TP / (TP + FN)

##### Precision
De los que ha predicho como 1, cuntos en realidad ha acertado. Precision, a diferencia del recall, pone foco en minimizar los FP. Como ejemplo podemos poner un filtro anti-spam. Que se cuele algn correo de spam (FN) no me importa. Ahora bien, si se clasifica como spam un correo importante del jefe (FP) y no lo leemos, s es ms grave.

Precision = TP / (TP + FP)

##### Specifity
Es el nmero de tems correctamente identificados como negativos sobre el total de negativos. Es lo opuesto al Recall.

Specifity = TN / (TN + FP)

##### F1-Score
Combinacin de las mtricas Precision y Recall. Esta mtrica se utiliza para comparar clasificadores, ya que es algo ms compleja de entender, pero muy til cuando tenemos clasificadores cuyos valores de recall y precision se intercalan unos con otros y no est muy claro cul es mejor, a no ser que tengamos claro que hay que centrarse o bien en el recall o en el precision. El rango del F1-score va de 0 a 1, como las mtricas anteriores.

F1-score = 2 * Precision * Recall / (Precision + Recall)


##### Precision vs Recall plot
Recuerda que el modelo por defecto calcula las probabilidades de pertenecer a una clase, y que por defecto, si por ejemplo el clasificador es binario, y la probabilidad cae por encima del 50% se clasificara como 1, como positivo. Ahora bien, modificar ese threshold cambia completamente la matrz de confusin. Si subimos el threshold somos ms restrictivos con los 1s, y por tanto empezar a clasificar ms valores como 0s y como FN. En cambio si lo bajamos, somos ms permisivos con los 1s, los clasificaramos casi todos, pero subiran los FP.

Es por ello que existen curvas como ROC o Precision vs Recall con las que podremos ver diferentes comportamientos del clasificador, teniendo en cuenta varios thresholds, y todo ello en una misma grfica.

Mediante esta curva obtendremos el punto ptimo de precision vs recall. Para usar esta grfica hay que comprender bien qu significan las mtricas de precision y de recall

<img src="./img/precs_recall.png" alt="drawing" width="500"/>

##### ROC Curve/AUC
Similar a la anterior curva. Nos sirve para comparar el recall con FPR = FP/(FP + TN)

De esta curva podemos obtener una medida, que es el AUC (Area Under the Curve). El AUC cuanto ms cercano a 1, mejor ser el clasificador. Si la curva es una lnea recta estaremos ante un clasificador muy malo, que no sera muy diferente a un clasificador aleatorio.

**Normalmente se usa la ROC curve cuando tenemos datasets balanceados, mientras que la curva precision-recall es ms propia de datasets con una clase mayoritaria.**

<img src="./img/roc.png" alt="drawing" width="500"/>

### 16.2 Mtricas para regresin

##### R^2
El R^2 es el coeficiente de determinacin, representa el porcentajede variacin de la variable dependiente vs la independiente para un modelo lineal. Se mide entre 0 y 1. El problema es que R^2 no nos indica si mi modelo se ajusta bien a mis datos. Un buen modelo puede tener un R^2 bajo, e incluso un modelo mal entrenado podra llegar a tener un R^2 alto. [En este artculo tienes ms detalles](https://statisticsbyjim.com/regression/interpret-r-squared-regression/#:~:text=R%2Dsquared%20evaluates%20the%20scatter,around%20the%20fitted%20regression%20line.&text=For%20the%20same%20data%20set,that%20a%20linear%20model%20explains.)

<img src="./img/r_squared.png" alt="drawing" width="230"/>

##### Mean Squared Error (MSE)
La media de los errores al cuadrado. Por qu al cuadrado? Porque al calcular la suma no queremos que se nos cancelen unos errores con otros. Qu est al cuadrado nos sirve como medida de errores? Perfectamente, es ms, esta mtrica magnifica los errores grandes, por lo que si esos son los errores que no corregimos, seguiremos teniendo un MSE alto. 

MSE es una mtrica muy utilizada ya que nos sirve para hacer foco en los errores grandes. Como los datos estn al cuadrado, no suele ser una mtrica para presentar resultados ya que los datos estn "distorsionados". **Se utiliza mayormente para comparar modelos**.

<img src="./img/mse.jpg" alt="drawing" width="200"/>

##### Root Mean Squared Error (RMSE)
Igual que en el caso anterior, pero ahora en unidades del target. Esta mtrica nos servir tanto para comparar modelos, como para presentar resultados.

<img src="./img/rmse.png" alt="drawing" width="180"/>

##### Mean Absolute Error (MAE)
Muy util cuando queremos medir errores en las unidades de la variable. Te dice qu error puedes esperar en la prediccin de la variable. Muy bueno cuando hay outliers, ya que lo vamos a ver claramente reflejado en la medida.

<img src="./img/mae.png" alt="drawing" width="200"/>

##### Mean Absolute Percentage Error (MAPE)
Es la nica mtrica acotada entre 0 y 1. Proporciona, en media, el % de error para cada punto respecto a su true label. Cuando queremos comparar modelos con diferentes targets, como por ejemplo en series temporales, las otras mtricas no nos sirven porque van en funcin de las unidades de cada target, sin embargo mediante el MAPE podremos comparar con la misma unidad de medidas.

<img src="./img/mape.jpeg" alt="drawing" width="230"/>


$$MAE = \frac{1}{n} \sum_{n=1}^{N}|y_i - \hat y_i|$$

$$SS_{res} = \sum_{i=1}^{N}(y_i - \hat y_i)^2$$

````python
# Metricas de regresin
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression

diamonds_df2 = pd.get_dummies(diamonds_df)

lr = LinearRegression()
X = diamonds_df2.drop(columns = ['price'])
Y = diamonds_df2['price']

lr.fit(X, Y)
preds = lr.predict(X)

print("Score del modelo (R^2):", round(lr.score(X, Y), 4))
print("R^2 score:", round(r2_score(preds, Y), 4))
print("MAE score:", round(mean_absolute_error(preds, Y), 4))
print("MSE score:", round(mean_squared_error(preds, Y), 4))
print("RMSE score:", round(np.sqrt(mean_squared_error(preds, Y)), 4))

def mean_absolute_percentage_error(y_pred, y_true): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100


print("MAPE score:", round(mean_absolute_percentage_error(preds, Y), 4))
````
